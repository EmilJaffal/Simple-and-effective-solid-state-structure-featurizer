```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
import collections

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

#Creates an dataframe NewElem that contains the values from Oliynyk_Element_Properties_Dataset
NewElem = pd.read_excel('element_properties_for_ML.xlsx') #, engine='openpyxl')
NewElem = NewElem.replace({"^\s*|\s*$":""}, regex=True) #Remove whitespaces in rows
NewElem.columns = NewElem.columns.str.replace(' ','') #Remove whitespaces in columns
NewElem['Symbol'].str.len() # check for whitespaces
NewElem = NewElem.reset_index() #Reset the indexes 
NewElem.columns = NewElem.columns.str.replace('[\n]', '') 
NewElem.index = NewElem.index + 1
NewElem = NewElem.fillna(0)
NewElem

#Test formulas 

List_A = ["Sc", "Y", "La", "Ce", "Pr", "Nd", "Sm", "Eu", "Gd", "Tb", "Dy", "Ho", "Er", "Tm", "Yb", "Lu", "Th", "U"]
List_B = ["Si", "Ga", "Ge", "In", "Sn", "Sb"]

composition = ["NdSi2", "Th2Os", "Sn5Co2", "Sm5Co7", "FeSb2", "Lu3Co2", 'SiC', 'C6H12']

df = pd.DataFrame(composition, columns=["Entry formula"])

Element1 = []
Composition2 = []
strings = list(df['Entry formula'])
for s in strings:
    Composition2.append(list(map(lambda x:'1' if x=="" else x,re.findall(r"[A-Z][a-z]*\s*([-*\.\d]*)", s))))
    Element1.append(re.findall(r"[A-Z][a-z]*\s*", s))  
	
for i in range(0, len(Element1)):
    if Element1[i][0] not in List_A:
        Element1[i][0], Element1[i][1] = Element1[i][1], Element1[i][0]
        Composition2[i][0], Composition2[i][1] = Composition2[i][1], Composition2[i][0]
		
		
df1 = pd.DataFrame(Element1, columns=['Element A', 'Element B'])
df2 = pd.DataFrame(Composition2, columns=['Index_A', 'Index_B'])
df2 = df2.apply(pd.to_numeric, errors='coerce')
df2.reset_index(drop=True)


df2['Normalized_Index_A'] = df2['Index_A'] / df2['Index_A'].max()
df2['Normalized_Index_B'] = df2['Index_B'] / df2['Index_B'].max()

main_df = pd.concat([df, df1], axis=1)
main_df = main_df.reset_index(drop=True)
main_df = pd.concat([main_df, df2], axis=1)
main_df = main_df.reset_index(drop=True)
main_df

main_df['A_AbundanceinEarthsCrust'] = main_df['Element A'].map(NewElem.set_index('Symbol')["AbundanceinEarth'scrust"])
main_df['B_AbundanceinEarthsCrust'] = main_df['Element B'].map(NewElem.set_index('Symbol')["AbundanceinEarth'scrust"])
main_df['A_Abundanceinsolarsystem'] = main_df['Element A'].map(NewElem.set_index('Symbol')["Abundanceinsolarsystem(log)"])
main_df['B_Abundanceinsolarsystem'] = main_df['Element B'].map(NewElem.set_index('Symbol')["Abundanceinsolarsystem(log)"])
main_df['A_HHIproduction'] = main_df['Element A'].map(NewElem.set_index('Symbol')["HHIproduction"])
main_df['B_HHIproduction'] = main_df['Element B'].map(NewElem.set_index('Symbol')["HHIproduction"])
main_df['A_HHIreserve'] = main_df['Element A'].map(NewElem.set_index('Symbol')['HHIreserve'])
main_df['B_HHIreserve'] = main_df['Element B'].map(NewElem.set_index('Symbol')['HHIreserve'])
main_df['weighted normalized Abundance in Earth’s crust (per atom)'] = (main_df['Index_A'] * main_df['A_AbundanceinEarthsCrust']) + (main_df['Index_B'] * main_df['B_AbundanceinEarthsCrust'])  / (
                                                          main_df['Index_A'] + main_df['Index_B']) 
main_df['weighted normalized Abundance in solar system crust (per atom)'] = (main_df['Index_A'] * main_df['A_Abundanceinsolarsystem']) + (main_df['Index_B'] * main_df['B_Abundanceinsolarsystem'])  / (
                                                          main_df['Index_A'] + main_df['Index_B']) 
main_df['weighted normalized HHI production (per atom)'] = (main_df['Index_A'] * main_df['A_HHIproduction']) + (main_df['Index_B'] * main_df['B_HHIproduction'])  / (
                                                          main_df['Index_A'] + main_df['Index_B']) 
main_df['weighted normalized HHI reserve (per atom)'] = (main_df['Index_A'] * main_df['A_HHIreserve']) + (main_df['Index_B'] * main_df['B_HHIreserve'])  / (
                                                          main_df['Index_A'] + main_df['Index_B']) 

main_df['A_Atomicweight'] = main_df['Element A'].map(NewElem.set_index('Symbol')['Atomicweight'])
main_df['B_Atomicweight'] = main_df['Element B'].map(NewElem.set_index('Symbol')['Atomicweight'])
main_df['molar_mass'] = main_df['A_Atomicweight'] + main_df['B_Atomicweight']
main_df['A/B_AtomicWeight'] = main_df['A_Atomicweight'] / main_df['B_Atomicweight']

main_df['A_Atomicnumber'] = main_df['Element A'].map(NewElem.set_index('Symbol')['Atomicnumber'])
main_df['B_Atomicnumber'] = main_df['Element B'].map(NewElem.set_index('Symbol')['Atomicnumber'])
main_df['A/B_AtomicNumber'] = main_df['A_Atomicnumber'] / main_df['B_Atomicnumber']
main_df['A_PeriodNumber'] = main_df['Element A'].map(NewElem.set_index('Symbol')['Period']).astype(int, errors='ignore')
main_df['B_PeriodNumber'] = main_df['Element B'].map(NewElem.set_index('Symbol')['Period']).astype(int, errors='ignore')
main_df['A_group number'] = main_df['Element A'].map(NewElem.set_index('Symbol')['Group']).astype(int, errors='ignore')
main_df['B_group number'] = main_df['Element B'].map(NewElem.set_index('Symbol')['Group']).astype(int, errors='ignore')
main_df['A-B_group number'] = main_df['A_group number'] - main_df['B_group number']

main_df['A_Mendeleevnumber'] = main_df['Element A'].map(NewElem.set_index('Symbol')['Mendeleevnumber']).astype(int, errors='ignore')
main_df['B_Mendeleevnumber'] = main_df['Element B'].map(NewElem.set_index('Symbol')['Mendeleevnumber']).astype(int, errors='ignore')
main_df['A-B_Mendeleevnumber'] = main_df['A_Mendeleevnumber'] - main_df['B_Mendeleevnumber']

main_df['A_valence e total'] = main_df['Element A'].map(NewElem.set_index('Symbol')['valenceetotal']).astype(int, errors='ignore')
main_df['B_valence e total'] = main_df['Element B'].map(NewElem.set_index('Symbol')['valenceetotal']).astype(int, errors='ignore')
main_df['A-B valence e total'] = main_df['A_valence e total'] - main_df['B_valence e total']
main_df['A+B valence e total'] = main_df['A_valence e total'] + main_df['B_valence e total']
main_df['Weighted_A+B valence e total'] = main_df['A+B valence e total']*main_df['Index_A']*main_df['Index_B']
main_df['Weighted_Normalised_A+B valence e total'] = main_df['A+B valence e total'] / np.max(main_df['A+B valence e total'])


main_df['A_unpairedelectrons'] = main_df['Element A'].map(NewElem.set_index('Symbol')['unpairedelectrons']).astype(int, errors='ignore')
main_df['B_unpairedelectrons'] = main_df['Element B'].map(NewElem.set_index('Symbol')['unpairedelectrons']).astype(int, errors='ignore')
main_df['A-Bunpairedelectrons'] = main_df['A_unpairedelectrons'] - main_df['B_unpairedelectrons']
main_df['A+Bunpairedelectrons'] = main_df['A_unpairedelectrons'] + main_df['B_unpairedelectrons']
main_df['Weighted_A+Bunpairedelectrons'] = main_df['A+Bunpairedelectrons']*main_df['Index_A']*main_df['Index_B']
main_df['Weighted_Normalised_A+Bunpairedelectrons'] = main_df['A+Bunpairedelectrons'] / np.max(main_df['A+Bunpairedelectrons'])

main_df['A_Gilman number of e'] = main_df['Element A'].map(NewElem.set_index('Symbol')['Gilmanno.ofvalenceelectrons']).astype(int, errors='ignore')
main_df['B_Gilman number of e'] = main_df['Element B'].map(NewElem.set_index('Symbol')['Gilmanno.ofvalenceelectrons']).astype(int, errors='ignore')
main_df['A-B Gilman number of e'] = main_df['A_Gilman number of e'] - main_df['B_Gilman number of e']
main_df['A+B Gilman number of e'] = main_df['A_Gilman number of e'] + main_df['B_Gilman number of e']
main_df['Weighted_A+B Gilman number of e'] = main_df['A+B Gilman number of e']*main_df['Index_A']*main_df['Index_B']
main_df['Weighted_Normalised_A+B Gilman number of e'] = main_df['A+B Gilman number of e'] / np.max(main_df['A+B Gilman number of e'])

main_df['A_Zeff'] = main_df['Element A'].map(NewElem.set_index('Symbol')['Zeff']) 
main_df['B_Zeff'] = main_df['Element B'].map(NewElem.set_index('Symbol')['Zeff']) 
main_df['A-B_Zeff'] = main_df['A_Zeff'] - main_df['B_Zeff']
main_df['A+B_Zeff'] = main_df['A_Zeff'] + main_df['B_Zeff']
main_df['Largest_Zeff'] = main_df[['A_Zeff', 'B_Zeff']].max(axis=1)
main_df['Lowest_Zeff'] = main_df[['A_Zeff', 'B_Zeff']].min(axis=1)
main_df['Avg_Zeff'] = (main_df['A_Zeff'] + main_df['B_Zeff'])/2

main_df['A_ionization energy'] = main_df['Element A'].map(NewElem.set_index('Symbol')['Ionizationenergy(eV)']) 
main_df['B_ionization energy'] = main_df['Element B'].map(NewElem.set_index('Symbol')['Ionizationenergy(eV)']) 
main_df['A-B_ionization energy'] = main_df['A_ionization energy'] - main_df['B_ionization energy']
main_df['A+B_ionization energy'] = main_df['A_ionization energy'] + main_df['B_ionization energy']
main_df['Largest_ionization energy'] = main_df[['A_ionization energy', 'B_ionization energy']].max(axis=1)
main_df['Lowest_ionization energy'] = main_df[['A_ionization energy', 'B_ionization energy']].min(axis=1)
main_df['Avg_ionization energy'] = (main_df['A_ionization energy'] + main_df['B_ionization energy'])/2


main_df['A_Coord_Number'] = main_df['Element A'].map(NewElem.set_index('Symbol')['CN']).astype(int, errors='ignore')
main_df['B_Coord_Number'] = main_df['Element B'].map(NewElem.set_index('Symbol')['CN']).astype(int, errors='ignore')
main_df['A-BCoord_Number'] = main_df['A_Coord_Number'] - main_df['B_Coord_Number']

main_df['A_ratio n closest/CN'] = main_df['Element A'].map(NewElem.set_index('Symbol')['rationclosest/CN']) 
main_df['B_ratio n closest/CN'] = main_df['Element B'].map(NewElem.set_index('Symbol')['rationclosest/CN']) 
main_df['Largest_ratio n closest/CN'] = main_df[['A_ratio n closest/CN', 'B_ratio n closest/CN']].max(axis=1)
main_df['Lowest_ratio n closest/CN'] = main_df[['A_ratio n closest/CN', 'B_ratio n closest/CN']].min(axis=1)
main_df['Avg_ratio n closest/CN'] = (main_df['A_ratio n closest/CN'] + main_df['B_ratio n closest/CN'])/2


main_df['A_polyhedron_distortion'] = main_df['Element A'].map(NewElem.set_index('Symbol')['polyhedrondistortion(dmin/dn)']) 
main_df['B_polyhedron_distortion'] = main_df['Element B'].map(NewElem.set_index('Symbol')['polyhedrondistortion(dmin/dn)']) 
main_df['Largest_polyhedron_distortion'] = main_df[['A_polyhedron_distortion', 'B_polyhedron_distortion']].max(axis=1)
main_df['Lowest_polyhedron_distortion'] = main_df[['A_polyhedron_distortion', 'B_polyhedron_distortion']].min(axis=1)
main_df['Avg_polyhedron_distortion'] = (main_df['A_polyhedron_distortion'] + main_df['B_polyhedron_distortion'])/2

main_df['A_CIF radius'] = main_df['Element A'].map(NewElem.set_index('Symbol')['CIFradiuselement']) 
main_df['B_CIF radius'] = main_df['Element B'].map(NewElem.set_index('Symbol')['CIFradiuselement']) 
main_df['A/B_CIF radius'] = main_df['A_CIF radius'] / main_df['B_CIF radius']
main_df['A-B_CIF radius'] = main_df['A_CIF radius'] - main_df['B_CIF radius']
main_df['Avg_CIF radius'] = (main_df['A_CIF radius'] + main_df['B_CIF radius'])/2

main_df['A_Pauling,R(CN12)'] = main_df['Element A'].map(NewElem.set_index('Symbol')['Pauling,R(CN12)'])
main_df['B_Pauling,R(CN12)'] = main_df['Element B'].map(NewElem.set_index('Symbol')['Pauling,R(CN12)'])
main_df['A/BPauling,R(CN12)'] = main_df['A_Pauling,R(CN12)'] / main_df['B_Pauling,R(CN12)']
main_df['A-BPauling,R(CN12)'] = main_df['A_Pauling,R(CN12)'] - main_df['B_Pauling,R(CN12)']
main_df['Avg_Pauling,R(CN12)'] = (main_df['A_Pauling,R(CN12)'] + main_df['B_Pauling,R(CN12)'])/2

main_df['A_Pauling EN'] = main_df['Element A'].map(NewElem.set_index('Symbol')['Pauling,R(CN12)'])
main_df['B_Pauling EN'] = main_df['Element B'].map(NewElem.set_index('Symbol')['Pauling,R(CN12)'])
main_df['A/BPauling EN'] = main_df['A_Pauling EN'] / main_df['B_Pauling EN']
main_df['A-BPauling EN'] = main_df['A_Pauling EN'] - main_df['B_Pauling EN']
main_df['Largest_Pauling EN'] = main_df[['A_Pauling EN', 'B_Pauling EN']].max(axis=1)
main_df['Lowest_Pauling EN'] = main_df[['A_Pauling EN', 'B_Pauling EN']].min(axis=1)
main_df['Avg_Pauling EN'] = (main_df['A_Pauling EN'] + main_df['B_Pauling EN'])/2

main_df['A_Melting point'] = main_df['Element A'].map(NewElem.set_index('Symbol')['Meltingpoint,K']) 
main_df['B_Melting point'] = main_df['Element B'].map(NewElem.set_index('Symbol')['Meltingpoint,K']) 
main_df['A/B_Melting point'] = main_df['A_Melting point'] / main_df['B_Melting point']
main_df['A-B_Melting point'] = main_df['A_Melting point'] - main_df['B_Melting point']
main_df['Largest_Melting point'] = main_df[['A_Melting point', 'B_Melting point']].max(axis=1)
main_df['Lowest_Melting point'] = main_df[['A_Melting point', 'B_Melting point']].min(axis=1)
main_df['Avg_Melting point'] = (main_df['A_Melting point'] + main_df['B_Melting point'])/2

main_df['A_MartynovBatsanovEN'] = main_df['Element A'].map(NewElem.set_index('Symbol')['MartynovBatsanovEN'])
main_df['B_MartynovBatsanovEN'] = main_df['Element B'].map(NewElem.set_index('Symbol')['MartynovBatsanovEN'])
main_df['A-B_MartynovBatsanovEN'] = main_df['A_MartynovBatsanovEN'] - main_df['B_MartynovBatsanovEN']
main_df['A/B_MartynovBatsanovEN'] = main_df['A_MartynovBatsanovEN'] / main_df['B_MartynovBatsanovEN']
main_df['Largest_MartynovBatsanovEN'] = main_df[['A_MartynovBatsanovEN', 'B_MartynovBatsanovEN']].max(axis=1)
main_df['Lowest_MartynovBatsanovEN'] = main_df[['A_MartynovBatsanovEN', 'B_MartynovBatsanovEN']].min(axis=1)
main_df['Avg_MartynovBatsanovEN'] = (main_df['A_MartynovBatsanovEN'] + main_df['B_MartynovBatsanovEN'])/2
main_df['A_Density'] = main_df['Element A'].map(NewElem.set_index('Symbol')['Density,g/mL'])
main_df['B_Density'] = main_df['Element B'].map(NewElem.set_index('Symbol')['Density,g/mL'])
main_df['A-B_Density'] = main_df['A_Density'] - main_df['B_Density']
main_df['A/B_Density'] = main_df['A_Density'] / main_df['B_Density']
main_df['Largest_Density'] = main_df[['A_Density', 'B_Density']].max(axis=1)
main_df['Lowest_Density'] = main_df[['A_Density', 'B_Density']].min(axis=1)
main_df['Avg_Density'] = (main_df['A_Density'] + main_df['B_Density'])/2

main_df['A_Specific heat'] = main_df['Element A'].map(NewElem.set_index('Symbol')['Specificheat,J/gK'])
main_df['B_Specific heat'] = main_df['Element B'].map(NewElem.set_index('Symbol')['Specificheat,J/gK'])
main_df['A-B_Specific heat'] = main_df['A_Specific heat'] - main_df['A_Specific heat']
main_df['A/B_Specific heat'] = main_df['A_Specific heat'] / main_df['B_Specific heat']
main_df['Largest_Specific heat'] = main_df[['A_Specific heat', 'B_Specific heat']].max(axis=1)
main_df['Lowest_Specific heat'] = main_df[['A_Specific heat', 'B_Specific heat']].min(axis=1)
main_df['Avg_Specific heat'] = (main_df['A_Specific heat'] + main_df['B_Specific heat'])/2

main_df['A_Cohesive energy'] = main_df['Element A'].map(NewElem.set_index('Symbol')['Cohesiveenergy'])
main_df['B_Cohesive energy'] = main_df['Element B'].map(NewElem.set_index('Symbol')['Cohesiveenergy'])
main_df['A-B_Cohesive energy'] = main_df['A_Cohesive energy'] - main_df['B_Cohesive energy']
main_df['A/B_Cohesive energy'] = main_df['A_Cohesive energy'] / main_df['B_Cohesive energy']
main_df['Largest_Cohesive energy'] = main_df[['A_Cohesive energy', 'B_Cohesive energy']].max(axis=1)
main_df['Lowest_Cohesive energy'] = main_df[['A_Cohesive energy', 'B_Cohesive energy']].min(axis=1)
main_df['Avg_Cohesive energy'] = (main_df['A_Cohesive energy'] + main_df['B_Cohesive energy'])/2

main_df['A_Bulk modulus'] = main_df['Element A'].map(NewElem.set_index('Symbol')['Bulkmodulus,GPa'])
main_df['B_Bulk modulus'] = main_df['Element B'].map(NewElem.set_index('Symbol')['Bulkmodulus,GPa'])
main_df['A-B_Cohesive energy'] = main_df['A_Bulk modulus'] - main_df['B_Bulk modulus']
main_df['A/B_Cohesive energy'] = main_df['A_Bulk modulus'] / main_df['B_Bulk modulus']
main_df['Largest_Cohesive energy'] = main_df[['A_Bulk modulus', 'B_Bulk modulus']].max(axis=1)
main_df['Lowest_Cohesive energy'] = main_df[['A_Bulk modulus', 'B_Bulk modulus']].min(axis=1)
main_df['Avg_Cohesive energy'] = (main_df['A_Bulk modulus'] + main_df['B_Bulk modulus'])/2

main_df.to_excel('AntonDescriptorsForBinary.xlsx', index=False)
main_df
```
